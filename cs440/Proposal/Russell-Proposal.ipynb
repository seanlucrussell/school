## Introduction

Generative adversarial networks are a new thing in machine learning. They work by playing a sort of game. One network attempts to immitate a dataset, and the other network looks differentiate immitations generated by its opponent from genuine data.

Even though this is a very new architecture it has been used to do some pretty neat stuff. In the paper that introduced the concept networks were used to generate new handwriting samples and new human faces. [https://arxiv.org/pdf/1406.2661.pdf] 

Additional work has show some very cool properties of the networks that arise from this process. For instance, the input vectors to the generative portion of the model hold some sort of meaning. Another team working with a generative network trained on human faces showed that an input vector representing a smiling man minus a vector for a neutral faced man plus the vector for a neutral faced woman resulted in a smiling woman.[https://arxiv.org/pdf/1511.06434.pdf%C3%AF%C2%BC%E2%80%B0]

Anyhow, I was in machine learning last semester and worked on a project involving the classification of instruments from audio files. So for this project, I was thinking of working on a project along the same lines, however instead of classifying my intention is to train a network to generate brief snippets of music. 

I do think this project might be slightly ambitious, but since I have some prior knowledge from my previous project and since a lot of resources are available online, I think I can manage.

## Methods

In cs480: Machine Learning, my final project used fourier analysis on an audio waveform to simplify the input vector to a neural network in order to cope with inputs based on a time series. This time I would like to examine the applications of recurrent neural networks in working with time series.

In addition, I wish to move from classifying audio towards generating audio. I think that generative machine learning techniques have a lot of potentially useful applications in imitating human creativity. Many papers explain generative adversarial networks with the metaphor of an art forger and an inspector, and I want to see for myself how powerful this technique is.

I intend to start small. While I will put a little more thought into this as the project progresses, I'll first look to generate music from just one sort of instrument and genre. If that looks to be successful, then I'll try working on adding in more instruments from more genres. This seems like the sort of problem that could require massive compute power and a massive number of samples to encapsulate all types of music. Fortunately, this problem is somewhat scalable, so I'll start small and see how far I can take things.

I'm not entirely certain what tools I will use yet. I believe there are some tutorials online about using tenserflow with generative models. However I know that there are a lot of different machine learning packages available, so I'd like to do my due dillegence before deciding on one.
